```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Condizioni iniziali
T = 4*np.pi
N = 100
initial_condition = np.array([
    0,
    1,
    1,
    0
])
# Exact solution
exact_solution_u = initial_condition[0] + np.linspace(0,T,N)*initial_condition[2]
exact_solution_v = initial_condition[1] + np.linspace(0,T,N)*initial_condition[3]

plt.plot(exact_solution_u, exact_solution_v)
plt.show()
```
