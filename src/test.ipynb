```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import savemat
from geometry import Sphere, Torus
from ode_methods import RungeKutta, AdamsBashforth
from numerical_solver import MultiStepSolver, draw_solver_solution
```

# Geodesics on the sphere

```python
# Numerical solver hyperparameters
total_time = 2*np.pi
steps_number = 100

# Initial condition
initial_condition = np.array([
    0,  # u0
    1,  # v0
    1,  # u'0
    0   # v'0
])

# Geometric domain
geometry = Sphere()

# Numerical solution
solver = MultiStepSolver(geometry, RungeKutta(4), AdamsBashforth(4))
solution = solver.solve(initial_condition, total_time, steps_number)

# Export solution
savemat('./solutions/sphere.mat',{'geodesic_1':solution})

# Drawing
fig, ax = draw_solver_solution(geometry, solution, (0,2*np.pi,40), (0,np.pi,20))
ax.tick_params(axis='both', which='both', length=0, labelleft=False, labelbottom=False)
plt.show()
```

# Geodesics on the torus

```python
# Numerical solver hyperparameters
total_time = 4*np.pi
steps_number = 150

# Initial condition
initial_condition = np.array([
    0,
    0,
    np.cos(0.845*np.pi/2),
    np.sin(0.845*np.pi/2)
])
# Geometric domain
geometry = Torus(2.0,1.0)

# Numerical solution
solver = MultiStepSolver(geometry, RungeKutta(4), AdamsBashforth(4))
solution = solver.solve(initial_condition, total_time, steps_number)

# Export solution
savemat('./solutions/torus.mat',{'geodesic_1':solution})

# Drawing
fig, ax = draw_solver_solution(geometry, solution, (0,2*np.pi,40), (0,2*np.pi,20))
ax.view_init(elev=35, azim=-50)
ax.tick_params(axis='both', which='both', length=0, labelleft=False, labelbottom=False)
plt.show()
```
